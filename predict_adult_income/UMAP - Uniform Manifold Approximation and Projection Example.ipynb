```python
import shap

X, y = shap.datasets.adult()
display(X.head())
display(y)
```

```python
import matplotlib.pyplot as plt
import umap
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401 unused import

n_neighbors = 10
min_dist = 0.0
n_components = 3
metric = 'euclidean'

embedding = umap.UMAP(
    n_neighbors=n_neighbors, min_dist=min_dist, n_components=n_components,
    metric=metric).fit_transform(X, y)

display(embedding)
```

```python
plt.rcParams['figure.figsize'] = [13, 10]
```

```python
y_class = list(map(lambda x: 'b' if x else 'r', y))
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.scatter(embedding[:, 0], embedding[:, 1], embedding[:, 2], c=y_class)
```

```python
plt.scatter(embedding[:, 0], embedding[:, 1], embedding[:, 2], c=y_class)
```

```python
for i in range(n_components):
    selected_x = i % n_components
    selected_y = (i+1) % n_components
    print('{},{}'.format(selected_x,selected_y))
    plt.scatter(embedding[:, selected_x], embedding[:, selected_y], c=y_class)
    plt.show()
```
